#### 1.2 Define Functions

```python
def stationary_check(x):
    print("Dickey–Fuller test: p=%f" % adfuller(x)[1])
    if adfuller(x)[1] > 0.05:
        print('Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.')
    else:
        print('Reject the null hypothesis (H0), the data does not have a unit root and is stationary.')

    
def plot_seasonal_decompose(x):
    '''this function conducts stationary check and seasonal decomposition'''
    fig = seasonal_decompose(x).plot()
    fig.set_size_inches(16, 8)
    fig.suptitle('Seasonal Decomposation')
    plt.show()
    
def plot_acf_and_pacf(x, lags):
    '''this function plots acf and pacf to identify patterns'''
    plt.figure(figsize=(16, 8))
    ax = plt.subplot(211)
    # Plot the autocorrelation factor (ACF)
    plot_acf(x, lags=lags, ax=ax)
    ax = plt.subplot(212)
    # Plot the partial autocorrelation factor (PACF)
    plot_pacf(x, lags=lags, ax=ax)
    plt.tight_layout()
    plt.show()
    

def arima_after_a_month_forecast(test_start_date, btc_month):
    '''this function fits arima models for the walk forward forecast of the close price and plots the results'''
    X = np.log(btc_month.close)
    test_dates = X.index[X.index>=test_start_date]

    pred_mean = []
    pred_lower = []
    pred_upper = []
    for t in test_dates:
        # past data end date
        t0 = X.index[X.index.get_loc(t)-1]

        # fit arima model
        model = SARIMAX(X[:t0], order=(1, 1, 0)).fit()
        btc_month_dynamic = model.get_prediction(start=t, end=t, dynamic=True, full_results=True)
        pred_mean.append(np.exp(btc_month_dynamic.predicted_mean.values[0]))

        # take 95% confidence interval 
        pred_dynamic_ci = btc_month_dynamic.conf_int(alpha=0.05)
        pred_lower.append(np.exp(pred_dynamic_ci['lower close'].values[0]))
        pred_upper.append(np.exp(pred_dynamic_ci['upper close'].values[0]))

    # create prediciton df
    pred_df = pd.DataFrame({'date':test_dates, 'forecast':pred_mean, 'lower':pred_lower, 'upper':pred_upper})
    pred_df['actual'] = btc_month.close[pred_df['date'].values].values

    # plot forecast and 95% CI
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=btc_month.index, y=btc_month.close, name='Actual'))
    fig.add_trace(go.Scatter(x=pred_df.date, y=pred_df.lower, showlegend=False, marker_color='lightgrey'))
    fig.add_trace(go.Scatter(x=pred_df.date, y=pred_df.upper, name = '95% CI', marker_color='lightgrey', fill='tonexty'))
    fig.add_trace(go.Scatter(x=pred_df.date, y=pred_df.forecast, name='ARIMA Forecast', marker_color='red'))
    fig.update_layout(title='Bitcoin monthly close price forecast',
                          xaxis_title='Date',
                          yaxis_title='Close price (USD)',
                          paper_bgcolor='rgba(0,0,0,0)',
                          plot_bgcolor='rgba(0,0,0,0)'
                        )
    
    return fig
```

#### 3.1 Stationary Check and Seasonal Decomposation
__Stationary:__ Time series with trends, or with seasonality, are not stationary. A stationary time series will have no predictable patterns in the long-term. If a time series is stationary, it implies the lack of broad trends (changes in mean and variance over time) in the data.

The Dickey–Fuller test suggests that the daily close price is not stationary. Before fitting any time series models, a series of steps are needed to make the process stationary.

```python
### check monthly data  
plot_seasonal_decompose(btc_month.close)
stationary_check(btc_month.close)       
```

#### 3.2 Log transformation
If the data show variation that increases or decreases with the level of the series (non-constant variance), then the process is nonstationary and a transformation is often desirable to suppress the variance and make it stationary. Logarithmic transformation or power transformation (box-cox transformation) are often employed.

In this section, we apply the log transformation on the daily close price to to suppress the non-constant variance. However, the log of the close price is still non-stationary from the Dickey–Fuller test, which indicates the further differencing is needed.

```python
# add log of close price column
btc_month['log_close'] = np.log(btc_month.close)
stationary_check(btc_month.log_close)       
```

#### 3.3 Difference
Another pre-processing step is differencing the data (calculating sequentially xt−xt−1) until we get to a point where the series is stationary. The first order differencing is applied as it performs the best and removes the non-stationary pattern.

```python
# Seasonal differentiation (12 months)
stationary_check(btc_month.log_close.diff(12).dropna())       
# Seasonal differentiation (3 months)
stationary_check(btc_month.log_close.diff(3).dropna()) 
# Regular differentiation (1 months)
stationary_check(btc_month.log_close.diff(1).dropna())
plot_seasonal_decompose(btc_month.log_close.diff(1).dropna())
```

#### 3.4 Autocorrelation
Draw autocorrelation factor (ACF) and partial autocorrelation factor (PACF) plots to identify patterns in the data which is now stationary on both mean and variance.

ARIMA(p,d,q):
* spikes in ACF - p - AR parameter
* spikes in PACF - q - MA parameter

As the ACF and PACF plots shows, there are not many spikes in the plots outside the insignificant zone (shaded) so there may not be enough information available in the residuals to be extracted by AR and MA models.

```python
plot_acf_and_pacf(btc_month.log_close.diff(1).dropna(), 12)
```

#### 3.5 ARIMA

##### 3.5.1 FInd the best ARIMA model

```python
# Initial approximation of parameters
qs = range(0, 3)
ps = range(0, 3)
d = 1
params = list(it.product(ps, qs))

# Model Selection
results = []
best_aic = float("inf")
warnings.filterwarnings("error")
for param in params:
    try:
        model = ARIMA(btc_month.log_close, order=(param[0], d, param[1])).fit()
    except:
        print('bad parameter combination (p, q):', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
warnings.resetwarnings()
```

### 4 Forecasting close price after a month -  Machine Learning

#### 4.1 create tabular dataframe

```python
# keep close price, volume, and market value of current date
btc_roll = btc[['close','volume','market']].copy()
btc_roll['trend'] = (btc_roll.close > btc_roll.close.shift(1))

# add a column of the close price after a month
btc_roll['close_after_month'] = btc_roll.close.shift(-30)

# add columns of average close price, spread, trend, and average volume over the past x days
for n_days in [7,30,90,180]:
    btc_roll['volume_avg_past_'+str(n_days)] = btc_roll.volume.rolling(n_days).mean()
    btc_roll['close_avg_past_'+str(n_days)] = btc_roll.close.rolling(n_days).mean()
    btc_roll['close_spread_past_'+str(n_days)] = btc_roll.close.rolling(n_days).apply(lambda x: max(x) - min(x))
    btc_roll['close_trend_past_'+str(n_days)] = btc_roll.close.rolling(n_days).apply(lambda x: x[0] < x[-1])
    btc_roll['up_ratio_past_'+str(n_days)] = btc_roll.trend.rolling(n_days).mean()
btc_roll.dropna(inplace = True)

# add weekday and month indicator after a month
btc_roll['weekday'] = (btc_roll.index + dt.timedelta(days=30)).weekday
btc_roll['month'] = (btc_roll.index + dt.timedelta(days=30)).month
btc_roll = pd.concat([btc_roll.drop(['weekday','month'], axis=1), pd.get_dummies(btc_roll[['weekday','month']].astype(str), drop_first = True)], axis=1)
```

#### 4.2 data preprocessing

```python
# create X and y
X = btc_roll.drop(['close_after_month'], axis=1)
y = btc_roll['close_after_month']

# standardize numerical features
bool_features = X.columns[X.nunique() == 2].values.tolist()
num_features = X.columns[X.nunique() > 2].values.tolist()
scaler = StandardScaler()
X_scaled = pd.concat([pd.DataFrame(scaler.fit_transform(X[num_features]), index = X.index, columns=num_features),
           X[bool_features]], axis=1)
```

```python
# create cross validation folds (time series split)
test_start_dates = ['2018-1-31','2018-3-31','2018-5-31']
test_end_dates = ['2018-3-31','2018-5-31','2018-7-31']
test_periods = list(zip(test_start_dates, test_end_dates))
def ts_train_test_split(test_start_date, test_end_date, df_ = btc_roll):
    df = df_.reset_index(drop=False).copy()
    train_index = df.index[(df.date <= test_start_date)].values.tolist()
    if test_end_date is None:
        test_index = df.index[(df.date > test_start_date)].values.tolist()
    else:
        test_index = df.index[(df.date > test_start_date)&(df.date <= test_end_date)].values.tolist()
    return (train_index, test_index)
cv = [ts_train_test_split(test_start_date, test_end_date) for test_start_date, test_end_date in test_periods]
```

#### 4.3 ML Modelling - Grid Search CV

### 5 Deep Learning - LSTM

#### 5.1 Feature Engineering

```python
from keras.layers import Input, Dense, Bidirectional, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Model
from keras.callbacks import EarlyStopping

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# define params
params = {
    "batch_size": 20,  # 20<16<10, 25 was a bust
    "epochs": 100,
    "lr": 0.00010000,
    "time_steps": 60
}

TIME_STEPS = params["time_steps"]
BATCH_SIZE = params["batch_size"]
EPOCH_SIZE = params["epochs"]
```

```python
# Adding relative columns
def add_relative_columns(df_):
    df = df_.copy()
    day_diff = df['close'] - df['open']
    df['rel_close'] = day_diff / df['open']
    df['high_low_ratio'] = df['high'] / df['low']
    df['rel_high'] = df['high'] / df['close']
    df['rel_low'] = df['low'] / df['close']
    return df

btc_dl = add_relative_columns(btc)
btc_dl = btc_dl[['close','rel_close','high_low_ratio','rel_high','rel_low','volume','market']]
```

```python
# create train set
btc_dl = add_relative_columns(btc)

train_cols = ['close','rel_close','high_low_ratio','rel_high','rel_low','volume','market']
# train_cols = ["close","open","high","low","volume"]
btc_dl = btc_dl[train_cols]

forecast_start_date = '2018-7-31'
train_index, test_index = ts_train_test_split(forecast_start_date, None, btc_dl)
training_set = btc_dl.iloc[train_index][train_cols]

sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
# prepare train history frame
# trim train set to fit batch size
training_set = training_set[:-(len(training_set)%BATCH_SIZE)]

y_col_index = 0
X_train = []
y_train = []
for i in range(TIME_STEPS, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-TIME_STEPS:i])
    y_train.append(training_set_scaled[i, y_col_index])
X_train, y_train = np.array(X_train), np.array(y_train)

#X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
# prepare test history frame
testing_set = btc_dl.iloc[list(range(test_index[0]-TIME_STEPS, test_index[0]))+test_index][train_cols]
# trim test set to fit batch size
testing_set = testing_set[:-(len(testing_set)%BATCH_SIZE)]
testing_set_scaled = sc.transform(testing_set)

X_test = []
y_test = []
for i in range(TIME_STEPS, len(testing_set_scaled)):
    X_test.append(testing_set_scaled[i-TIME_STEPS:i])
    y_test.append(testing_set_scaled[i, y_col_index])
X_test, y_test = np.array(X_test), np.array(y_test)

#X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# train LSTM
regressor = Sequential()

regressor.add(LSTM(units = 100, return_sequences = True, input_shape=(TIME_STEPS, X_train.shape[2])))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 60, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 40))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 20, activation='relu'))
regressor.add(Dense(units = 1, activation='sigmoid'))

# optimizer = optimizers.RMSprop(lr=params["lr"])
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

regressor.fit(X_train, y_train, epochs = EPOCH_SIZE, batch_size = BATCH_SIZE)
```

```python
# model prediction
y_pred = regressor.predict(X_test)
y_pred = y_pred.flatten()

# convert the predicted value to range of real data
y_pred_org = (y_pred * sc.data_range_[y_col_index]) + sc.data_min_[y_col_index]
y_test_org = (y_test * sc.data_range_[y_col_index]) + sc.data_min_[y_col_index]
```
